### Clock iterator

```python
# function style
def ClockIterator(second=0):
    
    while True:
        minute, sec = divmod(second, 60) 
        yield '{:02}'.format(minute) +':'+ '{:02}'.format(sec)
    
        second += 1
```

```python
clock = ClockIterator()
for time in clock:
    print(time)
```

```python
# class style
class ClockIterator:
    def __init__(self):
        self.second = 0
    
    # has to create a new method bc cannot have generator inside __init__
    def generator(self):
        while True:
            minute, sec = divmod(self.second, 60)
            yield '{:02}'.format(minute) +':'+ '{:02}'.format(sec)
            self.second += 1

        
```

```python
clock = ClockIterator().generator()
for time in clock:
    print(time)
```
